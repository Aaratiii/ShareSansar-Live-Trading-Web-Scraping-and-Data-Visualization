```python
from bs4 import BeautifulSoup
import requests

url = BeautifulSoup('https://www.sharesansar.com/live-trading', 'html.parser')
soup = requests.get(url)
code = BeautifulSoup(soup.text,"lxml")
code
```

```python
code_table = code.table
code_table
```

```python
tags = code_table.find_all('tr')
tags
```

```python
data = []
for tag in tags:
  data.append(tag.text.replace('\n                                        ',''))

data
```

```python
data = []
for tag in tags:
  data.append(tag.text.split('\n                                        '))

data
```

```python
data = []
for tag in tags:
  data.append(tag.text.replace('                                        ',''))

data
```

```python
data = []
for tag in tags[1:]:
  data.append(tag.text.replace('                                        ','').split('\n\n'))

data
```

```python
data = []
for tag in tags[1:]:
  x = (tag.text.replace('                                        ','').split('\n\n'))
  data.append(x[1:])
data
```

```python
data = []
for tag in tags[1:]:
  x = tag.text.replace('                                        ','').split('\n\n')
  y = [i.replace('\n','') for i in x]
  data.append(y[1:])

data
```

```python
col = tags[0].text.split('\n')[1:-1]
col
```
